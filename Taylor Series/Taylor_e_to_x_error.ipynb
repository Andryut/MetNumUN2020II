```
x=1
suma=0
for n in range (1,10):
  suma+=(-1)**(n+1)*x**n/n
print ("taylor:",suma)
print("ln x+1 con x=1:",math.log(x+1))
```

```
import math

x = 1
epsilon = 10**(-2)

for k in range(1,1000):
  if x**k/k < epsilon:
    break
print(k)

e_to_x = 0
for i in range(1,k):
    print(i,' ',e_to_x, ' delta ', x**i/i)
    e_to_x +=(-1)**(i+1)*x**i/i


    
print('math.log(1+1) = ',math.log(1+1))
```

```
import math

x = 1
epsilon = 10**(-2)
delta = 1
e_to_x = 0
i = 1
while  epsilon <= abs(delta):
  delta = (-1)**(i+1)*x**i/i 
  e_to_x += delta
  i = i+1
print(i-1) 
print('log x+1= ',e_to_x)    
print('ln (1+1) = ',math.log(x+1))
```

```
import math

x = 1
epsilon = 10**(-4)

for k in range(1000):
  if x**(2*k+1)/math.factorial(2*k+1) < epsilon:
    break
print(k)

e_to_x = 0
for i in range(k):
    print(i,' ',e_to_x, ' delta ', x**(2*i+1)/math.factorial(2*i+1))
    e_to_x +=(-1)**(i)*x**(2*i+1)/math.factorial(2*i+1)


    
print('math.sen(1) = ',math.sin(1))
```

```
import math

x = 1
epsilon = 10**(-4)
delta = 1
e_to_x = 0
i = 0
while  epsilon <= abs(delta):
  delta = (-1)**(i)*x**(2*i+1)/math.factorial(2*i+1)
  e_to_x += delta
  i = i+1
print(i-1) 
print('sin x= ',e_to_x)    
print('sin (1) = ',math.sin(x))
```
